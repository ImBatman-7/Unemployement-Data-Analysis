Which Months in Year 2020 have the highest unemployment rate in the country?

```python
df_11 = pd.read_csv("C:\\Users\\Manoj\\Desktop\\datasets\\Unemployment_Rate_upto_11_2020.csv")
df_11.head()
```

```python
# df_11["Month"] = df_11["Date"].dt.month
df_11.columns = [col.strip() for col in df_11.columns]
df_11.isnull().sum()
```

```python
df_11["Date"] = pd.to_datetime(df_11["Date"])
df_11.info()
```

```python
import calendar

df_11["Month"] = df_11["Date"].dt.month
df_11["Month"] = df_11["Month"].apply(lambda x : calendar.month_abbr[x])
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct']
df_11['month_order'] = pd.Categorical(df_11['Month'], categories=month_order)
df_11.head()
```

```python

df_mon = df_11.groupby("month_order").apply(lambda x : x.iloc[:,3].mean().round(2)).reset_index(name="EUR%")
df_mon = df_mon.sort_values(by="month_order")
df_mon
```

```python
plt.figure(figsize=(8,6)) # whole region 16x12
sbn.set_theme(font_scale=1) #increase font by 30%
axx = sbn.barplot(data=df_mon,x="month_order",y='EUR%')
axx.bar_label(axx.containers[0],fmt='%.2f') #labels in the bar, .2f means 2 decimal points
axx.set_xlabel("Month")
plt.show()
```

What is Labour Participation rate and Unemployement rate in each month?

```python
df_11_El = df_11.groupby("month_order")[["Estimated Unemployment Rate (%)","Estimated Labour Participation Rate (%)"]].mean().round(2)
df_11_El
```

```python
x1 = df_11["month_order"]
y1 = df_11_El["Estimated Unemployment Rate (%)"]
x2 = df_11["month_order"]
y2 = df_11_El["Estimated Labour Participation Rate (%)"]

fig = go.Figure()
fig.add_trace(go.Bar(x=x1,y=y1))
fig.add_trace(go.Bar(x=x2,y=y2))
fig.show()
```
